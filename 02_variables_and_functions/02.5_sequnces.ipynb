### Common operation in sequence


```python
#Check the length of sequence
list = [1,2,4,5,6,7,8]
print(len(list))  # len() count the total number of items present in it
```

```python
# membership list
items = ["pen", "eraser", "book"]
print("book" in items)
print(7 in items)

print("pen" in items)
print(3 not in items)
```

```python
#concatenation
list = [1,2,3]
list_2 = [4,5,6]
combined = list + list_2
print(combined)
```

```python
#Repitation
my_list = ["pen", "book", "eraser"]
rep_list = my_list*2
print(rep_list)
```

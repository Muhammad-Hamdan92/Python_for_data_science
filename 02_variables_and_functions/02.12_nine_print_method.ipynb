# Nine print() method

```python
# basic methiod
print("AOA, everyone") 

print("AOA", "everyone") # does not write comma

print("AOA", "everyone", sep = ",") # sep function use to write comma

print("AOA", "everyone", sep = ", ")

print("AOA", "everyone", sep = ", ", end = "!\n")
```

## Formated strings letrals
> this was introduced after python 3.6

```python
name = "Hamdna"
age = 15
question = "What are you doing, now"
print(f"My name is {name} and {age} years old.{question}") 
print(f"My name is {name} and {age} years old.\n{question}") #]n is used to break the line
```

## str.format method

```python
print("AOA, {}!".format("Hamdan"))
```

```python
print("My name is {0} and {1} years old. Currently, I am pursuing the Degree of {2}".format("Hamdan", 25, "Bsc.(Hons).Microbiology"))
```

## Percent % formationg
>It is an older method and simmilar to C-style method or string formating

```python
name = "Hamdan"
age = 17
print("My name is %s and %d years old." %(name, age)) #%s is for string and %d for integer
```

## sys.stdout.write formating
>unlike print(), allow you to write without adding a newline. It also give the length of string.

```python
import sys
sys.stdout.write("Hamdan! where are you going")
```

## Multi line print formating

```python
print("""AOA Hamdan,
What are you doing""")

# f-string with multi line printing
name = "Hamdan"
age = 17
print (f"""My name is {name} and {age} years old.
Who are you?""")
```

## Printing to file
>it help you to debugging the code in future.

```python
# print to file
with open("output.txt", "w") as file:
    print("Hello, Hamdan", file=file)
```

```python
# print to file to forward path
with open("./outputs/.print_to_file.txt", "w") as file:
    print("Hello, Hamdan", file=file)
```

# Indexing
>Accessing the single element from the ordered list.
Indexing is start from the 0\

#### Include:\
strings\
list\
tuples


```python
# string
name = "Hamdan"
print(name[0]) # give the first value
print(name[5])  # give the last value
print(name[-3]) # 3rd value from right to left 
```

```python
# list
fruits = ["apple", "banana", "pineapple"]
print(fruits[0]) # give the first value
print(fruits[2])  # give the last value
print(fruits[-3]) # 3rd value from right to left 
```

```python
#tuples
weigth = (12, 34, 45, 55, 89, 90)
print(weigth[4])
print(weigth[-4])
print(weigth[0])
```

```python
# matrix
my_matrix =[
        [1,0,0],
        [0,1,0],
        [0,0,1]
        ]

print(my_matrix)

print(my_matrix[2][0])
print(my_matrix[-1][2])
print(my_matrix[-2][2])
```

### Indexing error
>If the index is out of range than it give the error


```python
print(name[6])
len(name)                   #correct it 
print(my_matrix[3][-3])
len(my_matrix)
```

```python
# we can also use (1:2) if access more the values, but if the range is not match than it also give error
print(name[0:3])
print(name[1:5])  

print(name[0:]) #print everything

print(name[-1:-5]) # correct it {hint: indexing always from left to right}
```

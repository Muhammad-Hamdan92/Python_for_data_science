# Step 2: Handling missing value

````python
# df.drop()  remove all rows in which missig value are present

# df["prior_infection_site"].dropna() ## only remove the NA from this column but it also not the best way
df = df.drop(columns=['prior_infection_site'])# remove this column totaly and inplace=true is used to remove this column from main data
df
````

````python
# df.dropna(subset=["prior_infection_site", "resistance"])  ## also used to drop specific rows
````

````python
# df.fillna(0) ## this will not work because you can only impute in numeric vector or column
````

````python
df['resistance'] = df['resistance'].fillna(df['resistance'].mode()[0])# d  ## we used mode here because in the above code we were changed column into categorical

df['resistance'] = df['resistance'].astype(int) # convert back into integer
````

````python
df['resistance'].mean()
````

````python
df['resistance'] = df['resistance'].fillna(df['resistance'].median())  ## replace with median of resistance with NA in it
df
````

````python
df.isnull().sum()  ## completely impute and remove the na value
````

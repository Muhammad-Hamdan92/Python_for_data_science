# Subsetting datasets

```python
import pandas as pd

df = pd.read_csv('../00_datasets/microbial.csv')
df
```

```python
# subset on the basis of column
bacteria = df['bacteria']

print(type(bacteria))

bacteria
```

```python
# subset multiple column in dataset
df_bac_antibio = df[["bacteria", "antibiotic"]]

print(type(df_bac_antibio))
df_bac_antibio
```

```python
## subsetting with rows
import pandas as pd

K_pneu = df["bacteria"] == "K. pneumoniae"
K_pneu   ## if yo are using on single []
```

```python
import pandas as pd

K_pneu = df[df["bacteria"] == "K. pneumoniae"]
K_pneu   ## if yo are using on double []
```

```python
# subsetting with more than one condition
df_kleb_methi = df[
	(df["bacteria"] == "K. pneumoniae") &
	(df["antibiotic"] == "Methicillin")
]
df_kleb_methi
```

```python
##save the data
df_kleb_methi.to_csv('../00_datasets/0.1_pandas_practice_datasets/klebshela_methicillin', index=False)
```

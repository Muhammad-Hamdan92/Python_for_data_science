```python
import pandas as pd
from urllib.request import Request as req, urlopen  ## this is used to resolve the HTTP Error 403 , createS HTTP REQUEST

url = "https://en.wikipedia.org/wiki/World_population"

request= req(
    url,
    headers={"User-Agent": "for_webscraping_in_python/1.0"}  
)

html_url = urlopen(request)  #send request to wikipidia

tables = pd.read_html(html_url)

print(len(tables)) #print number of tables on web

df = tables[2] ## access the third tables on website
df
```

```python
##save the table as dataframe
df.to_csv('../00_datasets/0.1_pandas_practice_datasets/region_data_population.csv', index=False)
```

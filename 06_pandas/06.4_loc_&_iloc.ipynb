# loc and iloc function
>**loc** is used to specify rows and column on the basis of **rows and column labels**\
>**iloc** is used to specify rows and columns on the basis of **integer index**

## loc function

```python
import pandas as pd 

df = pd.read_csv('../00_datasets/microbial.csv')
df
```

```python
# df[df["bacteria", "antibiotic", "resistance"]]   ## same as below
df.loc[0:5, "bacteria":"resistance"] # here value after ':' will be included because it is labeled base
```

```python
# with negative indexing
# df.loc[-100:-50, 'bacteria' : 'resistance'] # this will not work because indexing is always [left to right]

df.loc[-50:100, 'bacteria' : 'resistance']
```

## iloc function

```python
df.iloc[0:5, 0:3]  #here the last one will excluded
```

# Types Of Casting

1. Int to Floating type [use float()]

```python
number = 4
float_num = float(number)
print(type(number))
print(type(float_num))
```

2. Float to Integer [use int()]

```python
float_age = 21.5
print(type(float_age))
int_age = int(float_age)
print(type(int_age))
```

3. Int/Float TO String[use str()]

```python
int_num = 10
float_num = 10.5
int_str = str(int_num)
float_str = str(float_num)
print(type(int_str))
print(type(float_str))
```

4. String to int/float [use(int()/float())]

```python
string_num = "10"
int_num = int(string_num)
string_float = "10.5"
float_num = float(string_float)
```

5. List to tuple/ tuple to List []

```python
my_list = [1, 2, 3, 4, 5]
print(type(my_list))
my_tuple = (1, 2, 3, 4, 5)
print(type(my_tuple))
```

```python
my_list_to_my_tuple = tuple(my_list)
print(type(my_list_to_my_tuple))

my_tuple_to_my_list = list(my_tuple)
print(type(my_tuple_to_my_list))
```

6. list to set 

```python
my_list = [1, 2, 3, 4, 5]
print(type(my_list))

my_list_to_my_set = set(my_list)
print(type(my_list_to_my_set))

again_to_my_list = list(my_list_to_my_set)
print(type(again_to_my_list))
```

7.   To boolean

```python
number = 10
print(type(number))

into_boolean = bool(number)
print(type(into_boolean))

empty_list = []
print(type(empty_list))

into_boolean_empty_list = bool(empty_list)
```

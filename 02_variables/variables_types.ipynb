# Vaiables types

1. Integer

```python
age = 21
print(type(age))
```

2. Floatings

```python
temperature = 36.6
print(type(temperature))
```

3. Strings

```python
greeting = "Hello, World!"
print(type(greeting))
```

4. Boolean

```python
is_classroom_full = False
print(type(is_classroom_full))
type(is_classroom_full)
```

5. List

```python
number = [2,5,6,7,8]
print(type(number))

person = ["name", "hamdan", "age", 30]
print(type(person))
```

6. Tuples

```python
tuple_example = (1, 2, 3)
print(type(tuple_example))
```

7. Dictionary

```python
person_info = {"name": "hamdan", "age": 30}
print(type(person_info))
```

8. Set

```python
person_set = {"hamdan", "ahmed", "ali"}
print(type(person_set))
```

9. None_type

```python
result = None
print(type(result))
```

10. Complex

```python
comple_x_number = 3 + 4j
print(type(comple_x_number))
```

# Level 2: **if-else condition**

# Level 3 **if-else-elif condition**

```python
score = 80    # change the value and see the output change
if score >= 80 and score <= 90:
    print("Grade A")
elif score >= 70 and score < 80:
    print("Grade B")
elif score >=60:
    print("Grade C")
else:
    print("Low grade")

vitus_load = 100000
if vitus_load > 50000:
    print("High load")
elif vitus_load > 25000:
    print("Medium load")
else:
    print("Low load")

humidity = 30
if humidity > 85:
    print("High humidity")
elif humidity >= 60:
    print("Moderate humidity")
elif humidity >= 45:
    print("Low humidity")
else:
    print("Dry")
```

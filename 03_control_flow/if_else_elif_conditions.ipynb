# With conditional operators
>There are five conditional operaters,following....

**==**   equal to

**!=**   not equal

**>**    greater than

**<**    smaller than

**>=**   greater than and equal to

**<=**   smaller than and equal to

# With logical operaters

**and**  condition will true if the both operands is true

**or**   condition will true if atleast one operands is true

**not**  condition is true if operands if false

```python
cgp = 3.5
skill = True
if cgp >= 3.5 and skill == True:
    print("got the scholarship")
```

```python
if cgp >= 3.5 or skill == false:
    print("may be got the scholarship")
```

```python
if not skill:
    print("got the scholarship")
else:
    print("never got the scholarship")
```

# Nested condition
>Here we can also utilize the conditional statement in nested mode e.g *if* inside the *if* condition or using the *elif* inside the any condition and at last *else* will use


```python
name = "Ali"
age = 20
if name == "Ali":
    if age >= 20:
        print("Ali is an adult")
    elif age <= 16 and age >= 13:
        print("Ali is a teenager")
    else:
        print("Ali is a child")
else:
    print("Ali is not the name")
```
